# LOAD ACCT table,  TACC_STATS dict, and some TACC Files

```python
acc_stats = pd.read_csv('./data/acc_stats/AccStats_Oct2014ToMar2015.tsv', delimiter='\t')
# acc_stats = acc_stats[[,'exec_host','start', 'end', 'JobID', 'Exit_status', 
#            'resources_used.walltime', 'resources_used.cput',
#           'resources_used.vmem', 'resources_used.mem']]


with open('./TACCStats_AttributeList.txt', 'rb') as attr:
    content = attr.readlines()
    content = content[4:]
    
TACC_HDR_TABLE = {}
for line in content:
    tacc_file_type, tacc_file_type_header = line.split('>>')
    tacc_file_type = re.sub('\d{2}\)\s', '', tacc_file_type).strip().strip('\\t')
    tacc_file_type_header = tacc_file_type_header.split(';')
    tacc_file_type_header = [i.strip() for i in tacc_file_type_header]
    TACC_HDR_TABLE[tacc_file_type] = tacc_file_type_header
    
# node_name_list = ['Node015.domain.name']
# base_path = './data/{}/'.format(node_name_list[0])
# tsv_name = 'intel_pmc3.tsv'
# tsv_abs_path = os.path.join(base_path, tsv_name)

def load_tacc_dataframe(tsv_abs_path):
    file_name = os.path.split(tsv_abs_path)[1]
#     print file_name
    df = pd.read_csv(tsv_abs_path,
                                   delimiter='\t',
                                   names=TACC_HDR_TABLE[file_name],
                                   index_col='Timestamp' )
    return df


```
